```python
from llama_parse import LlamaParse
from llama_index.core import SimpleDirectoryReader

from dotenv import load_dotenv

load_dotenv()

import nest_asyncio
nest_asyncio.apply()
```

```python
parser = LlamaParse(
    result_type="text"  # "markdown" and "text" are available
)

file_extractor = {".pdf": parser}
documents = SimpleDirectoryReader(input_files = ["Slides/lecture3.pdf"], file_extractor=file_extractor).load_data()
```

```python
from pprint import pprint
import os
```

```python
pprint(documents)
```

```python
file_name = documents[0].metadata['file_name']
file_name = file_name.split(".")[0]

for page_num, doc in enumerate(documents):
    
    txt_file_path = f"Parsed_Slides/{file_name}.txt"

    try:
        page_text = doc.text

        if page_num == 0:
            with open(txt_file_path, 'w', encoding='utf-8') as txt_file:
                txt_file.write(f"--- Page {page_num + 1} ---\n{page_text}\n\n")

        else:        
            with open(txt_file_path, 'a', encoding='utf-8') as txt_file:
                txt_file.write(f"--- Page {page_num + 1} ---\n{page_text}\n\n")

        print(f"Text extracted and saved to {txt_file_path}")
    
    except Exception as e:
        print(e)
```

```python
input_directory = "Slides"
pdf_files = [os.path.join(input_directory, file) for file in os.listdir(input_directory) if file.endswith(".pdf")]

# Load data for each file in the directory
for pdf_file in pdf_files:
    documents = SimpleDirectoryReader(input_files=[pdf_file], file_extractor=file_extractor).load_data()
    file_name = documents[0].metadata['file_name']
    file_name = file_name.split(".")[0]

    for page_num, document in enumerate(documents):
        
        txt_file_path = f"Parsed_Slides/{file_name}.txt"

        try:
            page_text = document.text

            if page_num == 0:
                with open(txt_file_path, 'w', encoding='utf-8') as txt_file:
                    txt_file.write(f"--- Page {page_num + 1} ---\n{page_text}\n\n")

            else:        
                with open(txt_file_path, 'a', encoding='utf-8') as txt_file:
                    txt_file.write(f"--- Page {page_num + 1} ---\n{page_text}\n\n")

            print(f"Text extracted and saved to {txt_file_path}")
        
        except Exception as e:
            print(e)
```
